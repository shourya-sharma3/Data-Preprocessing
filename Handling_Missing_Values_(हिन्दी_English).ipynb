We have two methods that we can use to deal with missing values in a dataset <br>
1. Imputation
2. Dropping

## **Imputation**
Replace Nan values with central tandencies like mean, median and mode (depends on data distribution and is later covered in notebook)

## **Dropping**
simply removing them

<br>
हमारे पास दो विधियाँ हैं जिनका उपयोग हम डेटासेट में गायब मानों (Missing Values) को संभालने के लिए कर सकते हैं।  
<br>  
1. इम्प्यूटेशन (Imputation)
2. ड्रॉपिंग (Dropping)  

## **इम्प्यूटेशन (Imputation)**  
NaN मानों को केंद्रीय प्रवृत्तियों (Central Tendencies) जैसे कि **माध्य (Mean), माध्यिका (Median), और बहुलक (Mode)** से प्रतिस्थापित करना।  
(यह डेटा वितरण पर निर्भर करता है और इसे आगे नोटबुक में कवर किया गया है।)  

## **ड्रॉपिंग (Dropping)**
साधारण रूप से गायब मानों वाली पंक्तियों को हटा देना।  
<br>

**Imputation** is generally done on small data where we cant remove null values<br>
**Dropping** is generally done with big data and reletively small null data points
<br>
**इम्प्यूटेशन (Imputation)** आमतौर पर छोटे डेटा पर किया जाता है, जहाँ हम नल (Null) मानों को हटा नहीं सकते।  <br>
**ड्रॉपिंग (Dropping)** आमतौर पर बड़े डेटा पर किया जाता है, जब नल मान अपेक्षाकृत कम होते हैं।  
<br>

# Imputation

We first analyse what tandency is the most suitable through data analysis<br>
हम पहले डेटा विश्लेषण (Data Analysis) के माध्यम से यह विश्लेषण करते हैं कि कौन सी केंद्रीय प्रवृत्ति (Tendency) सबसे उपयुक्त है।  
<br>

```python
import seaborn as sns
sns.displot(placementdata["salary"])
```

we can observe that data distribution is uneven because some of the data points go very high, which will in turn give us wrong information

Lets take an example
Say Person1, 2, 3, 4 have salary 10,20,30 and 40 respectively and the person 5 have 200, than the mean will not show accurate result in this situation <br><br>
हम देख सकते हैं कि डेटा वितरण असमान (Uneven) है क्योंकि कुछ डेटा बिंदु बहुत अधिक हैं, जिससे गलत जानकारी मिल सकती है।  

### **उदाहरण:**  
माना कि **व्यक्ति 1, 2, 3, 4** की सैलरी क्रमशः **10, 20, 30 और 40** है, लेकिन **व्यक्ति 5** की सैलरी **200** है।  

अब, यदि हम **माध्य (Mean)** निकालते हैं तो वह हमे गलत जानकारी देगी <br><br>
That is why Median is more prefered
<br >
इसलिए, इस प्रकार के डेटा वितरण में **माध्यिका (Median)** का उपयोग करना अधिक उपयुक्त होगा।  
<br>
